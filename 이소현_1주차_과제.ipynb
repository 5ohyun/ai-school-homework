# 1. 언어

가장 관심있게 본 인공지능 분야이며 인공지능을 사용하는 언어 서비스 및 제품들이 많이 있다. 
NLP라고 하여 자연 언어를 처리하는 기술로 불리고 번역, 챗봇과 같은 질의응답 기능 등이 가능하다. 또한 스팸 메일 필터링, 자동 요약, 감성분석까지 무궁무진하게 크는 분야이며
어마어마한 양의 텍스트 정보들로부터 의미있는 정보를 가져올 수 있다는 점이 굉장히 신기하게 다가왔다.

그 중 가장 놀랍게 봤던 서비스 중 하나는 당근마켓이다.지역 기반으로 중고거래를 할 수 있도록 도와주는 어플이며 급속도로 성장하고 있는 기업이다.
언어 분야에서 뚜렷하게 강세를 보이고 있는 것은 아니지만 일반적인 커뮤니티 어플이라고만 생각했던 나에게는 인공지능이라는 게 거창한 것이 아니라 이렇게 작은 부분들까지도 다양하게 쓰일 수 있다는 깨달음을 주었다.

당근마켓에서는 인공지능 언어기술 BERT를 사용해 동네 생활 게시글 필터링 모델을 구현하고 있다. 일반적인 중고거래 글이 올라오는 탭과 분리하여 일반 게시글이 올라오는 동네생활탭으로 구성하였는데 사람이 보고 동네 생활탭에 어울리는 글인지 아닌지를 분류하고 있었다고 한다.
하지만 이런 서비스를 전국적으로 운영하게 된다면 사람의 손을 다 거칠 수 있는 것이 아니므로, 게시글 필터링 기술을 사용해 동네생활에 맞는지 판별하는 딥러닝 모델을 구현했다.
자연어 처리 분야에서는 이런 토픽 모델링을 할 때 CNN이나 LSTM을 이용한 모델을 많이 사용하나 데이터가 적고 과적합 문제로 인해 2018년도에 나온 구글에서 발표한 NLP 모델인 BERT를 사용했다. BERT는 사람보다 더 답을 잘 맞춘다는 평이 있을 정도로 자연어 의미 파악 정확도가 높은 기술이라고 한다.

BERT는 영어 위주로 개발된 기술인데 ETRI 한국전자통신연구원에서 개발한 한국어 언어모델 KoBERT 도 존재한다.
한국어는 교착어라고 해서 형태소에 다른 형태소들이 붙어 단어가 구성되는데 이러한 한국어의 특징 때문에 자연어처리가 어렵다. KoBERT는 그런 특성까지 반영하여 만들어져 구글이 배포한 한국어 모델보다 높은 성능을 보이고 이를 바탕으로 법령분야에 대한 질의 응답 서비스와 유사한 특허를 분류하는 지능형 분석 기술도 준비하고 있다고 한다. 

또한 카카오에서는 딥러닝을 기반으로 한 형태소 분석기 '카이'를 오픈소스로 제공하고 있고 LG에서는 장문의 답변도 가능하도록 질문 응답하는 '코쿼드' 등 소기업부터 대기업까지 언어분야는 다양한 산업에서 활용되고 있다.

#2. 음성

음성 분야에서는 자동 번역, 자동 질의응답 서비스 등의 사례들이 있는데 인공지능 스피커가 가장 눈에 띄는 음성 인공지능 서비스이다.
요즘 스마트 스피커가 각 가정에 하나씩은 있을 정도로 시중에 많이 존재하며 실생활에 가장 잘 녹아있는 인공지능 분야가 아닐까 싶다. 처음에 삼성의 빅스비나 애플의 시리가 출시되었을 때 낮은 정확도에 실망하고 사용성이 떨어질 것이라고 예상하였으나, KT의 기가지니와 데이트를 하는 영상까지 나올 정도로 점차 대화까지 가능한 기술이 되었다.
이제는 외국어를 번역해주는 기술부터 사투리, 남녀노소 누구든지의 대화를 이해하고 해석하는 기술까지 계속 발전하고 있는 분야이다.

AI 스피커의 원리를 찾아보니, 사람의 목소리가 마이크를 통해 음성 신호로 입력되고, 그 신호를 음성 신호 벡터로 변화시킨다고 한다. 전체 음성 신호 벡터는 기존 단어 데이터 베이스에 따라 분할되어 언어 표현 정보로 추출되는 과정을 거치게 된다. 이 추출된 정보는 입력된 정보에 알맞는 반응을 보일 수 있도록 스킬을 결정하는 '스킬 분류기'에 입력되고, 스킬과 연결되어 의도를 파악하는 '의도 분류기'에도 입력된다. '의도 분류기'는 축적되는 데이터를 바탕으로 계속 학습한다. 또 정말 의미있는 정보가 무엇인지 인지하도록 문장의 문법적 구조를 고려해 핵심 정보를 슬롯으로 찾아 인지하는 '슬롯 인지기'에 입력된다.

사용자의 언어 습관, 자주 방문하는 사이트, 자주 듣는 노래 등 사용자의 특성과 취향에 따라 저장소가 변화하고 AI 스피커는 사용자의 명령을 더 정확하게 이해할 수 있도록 발전한다. 그리고 집 환경과도 연결되어 조명 등의 기기를 제어하는 기능부터 신나는 노래를 틀어달라고 말하면 재생해주는 노래추천기능, 음식 레시피 등이 궁금하면 대답해주는 질의응답기능, 날씨기능, 알람기능 등 이제는 정말 사람의 언어를 잘 이해하는 모습을 보인다. 이렇게 계속 학습하고 발전하는
음성 분야는 사람과 자연스럽게 일상대화까지 가능한 수준이 곧 될 것 같다.

#3. 이미지

요즘 공항에서는 사람이 일일이 보안 검색대에서 검사하는 것이 아니라 기계가 이미지를 보고 분석하는 기술을 사용하고 있다. 또한 휴대폰 앨범을 보면 어떤 인물인지 음식인지 동물인지 자동으로 분류를 하고 있으며 카메라에 꽃을 비추면 어떤 꽃인지 알아맞추기까지 한다. 

이미지를 통해 사물을 분류하는 구글 렌즈 서비스에 대해 알아보았는데 이 구글 렌즈는 4가지 기술을 가지고 있다. 이미지 인식은 이미지의 특성을 잘 포착하고 추상화하는 것이 중요한데 여러 개의 사물을 분류해 내 사물에 태그를 붙이는 분류 기술, 이미지의 위치를 파악하는 발견 기술, 각 이미지의 유사도를 판단하고 같은 사물이지만 다른 각도에서 찍힌 것인지 등을 벡터값으로 판단하는 임베딩 기술, 연산 속도의 개선 기술을 통해 이미지 분석 분야에서 많은 성과를 이루고 있다.

이미지를 분석할 때 컴퓨터는 픽셀로 수많은 신경망 계층을 거쳐 사람처럼 인식할 수 있도록 딥러닝을 사용하며 사진을 보고 사물의 벡터값을 만들어 패턴을 파악해 패션 분야 등에서도 유사한 제품을 빠르게 찾아주는 사례로 쓰이고 있다고도 한다.

이미지 분류를 위한 알고리즘에는 NN(nearest neighbor), KNN(k-nearest neighbor), CNN (Convolutional neural network)등 다양하게 있으며 CNN은 특히 영상 처리에 특화 되어 있다.


# 4. 자율주행

자율 주행 기술은 사람이 운전할 때 눈으로 각 차선을 인식하기, 다른 차들을 살펴보기, 신호, 건물 등 파악하기, 적절한 속도와 방향 등을 정하는 행동을 수행하는 역할을 한다.

자율주행차는 운전자의 행동을 스스로 하는 차로 운전자가 관여하지 않아도 인지하고 판단하며 제어해야한다. 처음으로 인지하는 과정에서는 디지털 지도를 통해 경로탐색을 하고, 다른 차량들과의 통신, 교통 인프라 통신을 통해 빠르게 도로 상황을 파악한다. 두번째, 판단하는 과정에서는 차선 변경, 추월, 정차 등의 주행 상황을 판단하고 결정한다. 세번째, 제어하는 과정에서는 엔진을 가속할 것인지 감속할 것인지 등의 역할을 한다.

모든 과정들은 순차적이 아닌 통합적이고 유기적인 단계이며 동시에 일어나야하며 특히 판단 및 제어 단계에서는 보행자 등 돌발상황 등에 대처하는 능력이 필요하며 다양한 센서를 통해 정보를 수집되어야 한다.

자율 주행 회사로 뜨거운 인기를 보이고 있는 테슬라는 완전 자율 주행이 가능한 차량을 출시한다고 밝혔고 상용화가 가능해지고 있는 분야이다. 인간이 감지할 수 없는 범위까지 살펴보며 8개의 카메라, 12개의 센서 등이 사용되었다고 한다.
